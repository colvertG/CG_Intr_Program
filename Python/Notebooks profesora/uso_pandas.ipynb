## Procesamiento de datos


```python
# Show the data types for each column
pacientes.dtypes
```

```python
# get the mean 
pacientes['age'].mean()
```

```python
# get the median 
pacientes['age'].median()
```

```python
# get the mean grouped by sex
pacientes[['sex', 'age']].groupby('sex', as_index=False).mean()
```

```python
# get the median grouped by sex
pacientes[['sex', 'age']].groupby('sex', as_index=False).median()
```

```python
# get the median grouped by type of sex and met_s
pacientes[['sex', 'met_s', 'age']].groupby(['sex', 'met_s'], as_index=False).median()
```

```python
# Filtrado
# get all rows with BMI < 24.9
pacientes_BMI_normal = pacientes[pacientes['BMI'] < 24.9]
pacientes_BMI_normal.head()
```

```python
# get all rows with BMI < 24.9 and met_s to 1
pacientes_BMI_normal_met_s = pacientes[(pacientes['BMI'] < 24.9) & (pacientes['met_s'] == 1)]
pacientes_BMI_normal_met_s.head()
```
